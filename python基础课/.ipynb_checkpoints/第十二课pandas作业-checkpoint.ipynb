```python
titanic = pd.read_csv("seaborn-data/titanic.csv")
```

```python
titanic.head()
```

```python
newcolumns = list(titanic.columns)
newcolumns.remove("age")
newcolumns
```

```python
titanic1 = titanic[newcolumns]
titanic1
```

```python
Age = titanic["age"]
Age
```

```python
def age_update(age):
    if age < 10:
        return "0s"
    elif age < 20:
        return "10s"
    elif age < 60:
        return "50s"
```

```python
Age = Age.apply(age_update) # 不能对原数组修改
Age
```

```python
titanic.groupby(["sex",Age])["survived"].mean().unstack()
```

```python
titanic.pivot_table("survived",index="sex",columns=Age)
```
