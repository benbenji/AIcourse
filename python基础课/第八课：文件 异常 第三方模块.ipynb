# 文件

## 文件的读写

```python
with open("test.txt","r",encoding= "utf8") as f:
    text = f.read()
    print(text)
```

路径可以是相对路径，也可以是完整路径

## 打开模式
* 'r' 以只读方式打开文件（默认）
* 'w' 以写入的方式打开文件，会覆盖已存在的文件
* 'x' 如果文件已经存在，使用此模式打开将引发异常
* 'a' 以写入模式打开，如果文件存在，则在末尾追加写入
* 'b' 以二进制模式打开文件
* 't' 以文本模式打开（默认）
* '+' 可读写模式（可添加到其他模式中使用）
* 'U' 通用换行符

* readline

```python
with open("test.txt","r",encoding="utf8") as f:
    while True:
        line = f.readline()
        if line:
            print (line)
        else:
            break
```

## 异常处理
1. try except

```python
x = 10
y = 0
try:
    z = x / y
except ZeroDivisionError:
    x = x /(y+1e-7)
    print("0不可以做除数")
 
```

2. 多分支

```python
ls = []
d = {"name":"大眼仔"}
try:
    d["age"]
except NameError:
    print("变量名不存在")
except IndexError:
    print("超出限制")
except KeyError:
    print("键值不存在")
```

* 万能异常

```python
ls = []
d = {"name":"大眼仔"}
try:
    d["age"]
except Exception:
    print("出错了")
```

```python
ls = []
d = {"name":"大眼仔"}
try:
    d["age"]
except Exception as e:
    print(e)
```
