```python
class CreditCard:
    def __init__(self,Name,Amount,CurrentAmount,SwipOneLimit):
        self.Name = Name
        self.Amount = Amount
        self.CurrentAmount = CurrentAmount
        self.SwipOneLimit = SwipOneLimit
    def getName(self):
        return self.Name
    def getAmount(self):
        return self.Amount
    def getCurrentAmount(self):
        return self.CurrentAmount
    def getSwipOneLimit(self):
        return self.SwipOneLimit
    def setAmount(self,amonut):
        self.Amount = amonut
    def setSwipOneLimit(self,limit):
        self.SwipOneLimit = limit
    def SwipCard(self,money):
        if (self.Amount - money) < 0:
            print("超出额度")
            return False
        elif money > self.SwipOneLimit:
            print("超出单次刷卡额度")
            return False
        else:
            self.CurrentAmount -= money
            return True
```

```python
class ChinaBankCreditCard(CreditCard):
    def __init__(self,Name,Amount,CurrentAmount,SwipOneLimit,shopsList,score):
        super().__init__(Name,Amount,CurrentAmount,SwipOneLimit)
        self.score = score
        self.shopsList = shopsList
    def SwipCard(self,money,shop):
        if shop in self.shopsList:
            money = money * 0.95
        if super().SwipCard(money):
            self.score += money // 10
```

```python
mycard = ChinaBankCreditCard("benbenji",1000,100,30,["haha","xingbake"],0)
```

```python
mycard.SwipCard(20,["haha"])
```
