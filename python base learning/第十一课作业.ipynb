```python
A = np.random.random(size = (3,3))
B = np.random.random(size = (3,3))
```

```python
%%time
ABfor = np.zeros([3,3])
for i in range(3):
    for j in range(3):
        result = 0
        for k in range(3):
            result+=A[i][k]*B[k][j]
        ABfor[i][j] = result
```

```python
ABfor
```

```python
%%time
ABfor1 = np.dot(A,B)
```

```python
ABfor1
```

```python
%%time
mulABfor = np.zeros((3,3))
for i in range(3):
    for j in range(3):
        mulABfor[i][j] = A[i][j]*B[i][j]
mulABfor
```
