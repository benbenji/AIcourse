```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10,10,1000)
y = 1/(1 + np.exp(-x))
y1 = (np.exp(x) - np.exp(-x))/(np.exp(x) + np.exp(-x))
with plt.style.context("seaborn-ticks"):
    plt.plot(x,y1,"g-",linewidth=2,label="Curve 1")
    plt.plot(x,y,"r:",linewidth=3,label="Curve 2")
    plt.title("Two curves")
    plt.xlim(-10,10)
    plt.ylim(-1.5,1.5)
    plt.xticks(np.arange(-10,10,step=1),fontsize=15)
    plt.legend()
    plt.savefig("Two_curves.png")
```

```python
x = np.linspace(-10,10,1000)
y = 1/(1 + np.exp(-x))
y1 = (np.exp(x) - np.exp(-x))/(np.exp(x) + np.exp(-x))
y2 = np.sin(x)
y3 = np.cos(x)
```

```python
plt.subplot(221)
plt.plot(x,y,'o-',markerfacecolor="r", markersize=1)
plt.subplot(222)
plt.plot(x,y1,"r--")
plt.subplot(223)
plt.plot(x,y2,"b+")
plt.subplot(224)
plt.plot(x,y3,"g-")
```

```python
mu, sigma = 100, 5
x = mu + sigma * np.random.randn(10000)
plt.hist(x, bins=50, #区间
         facecolor='g', alpha=0.75)
```

```python
from scipy.stats import norm
_, bins, __ = plt.hist(x,50,density=True) #获取
y = norm.pdf(bins,mu,sigma) #返回一个区间的概率密度
plt.plot(bins,y)
```

```python
_, bins, __ = plt.hist(x,50,density=True, cumulative=True) 
y = norm.cdf(bins,mu,sigma)
plt.plot(bins,y,'r--',lw=3)
```
