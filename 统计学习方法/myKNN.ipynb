1. 思考k近邻算法的模型复杂度体现在哪里？什么情况下会造成过拟合？打卡进行文字说明。
    算法复杂度体现在样本的数量，数量越多计算代价越高，过拟合主要当k值取得小时，会导致模型过于集中考虑局部信息，无法充分获取全局的估计

```python
import numpy as np
def load_data():
    x = np.array([[5, 4],[9, 6],[4, 7],[2, 3], [8, 1], [7, 2]])
    y = np.array([1, 1, 1, 0, 0, 0])
    return x, y
X, Y = load_data()
```

```python
def KNN(test, X ,Y):
    dif = X - test
    dis = np.sum(dif**2, axis=1)
    return Y[np.argmin(dis)]

%timeit KNN((5,3), X, Y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=1)
neigh.fit(X, Y)
%timeit neigh.predict([[5, 3]]) #我的比他快呀

```
